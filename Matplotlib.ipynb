# Matplotlib provides two main ways to plot:
1. StateFull Interface - Pyplot
2. Object-Oriented Interface - Manually Created



### A. Stateful (Pyplot) Interface
- Easy and quick
- Uses the `matplotlib.pyplot` module (usually imported as `plt`)
- Automatically tracks the "current" figure and axes

📄 Docs: [matplotlib.pyplot — Pyplot tutorial](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)

---

### B. Object-Oriented (OO) Interface
- More flexible and powerful
- You create and manage the `Figure` and `Axes` objects explicitly
- Preferred for complex/multi-plot figures

📄 Docs: [Object-Oriented Interface Overview](https://matplotlib.org/stable/users/explain/figure/intro.html)


### ✅ Pyplot Interface

When you do:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```
Matplotlib will:

Check if a figure exists → if not, it creates one.

Check if an axes exists → if not, it adds one.

Then it plots onto that axes.

✅ Good for quick plots
❌ Not modular or scalable for bigger applications.

### ✅ Object-Oriented Interface Interface

When you do:
```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
plt.show()
```
Here:

fig is a Figure object.

ax is an Axes object (or array of them, if using multiple subplots).

✅ You control everything: layout, axes, styling.

✅ Recommended for most non-trivial programs.



# Example Pyplot Interface


````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [1, 4, 9])  # Auto-creates figure and axes
plt.title("Pyplot Interface") #set the title of the plot
plt.xlabel("x") #set the name of x-axis
plt.ylabel("y") #set the name of y-axis
plt.grid(True) #set grid like structure (grid can be visible) - deafult its False
plt.show() #show the plot
````

# Example Object Oriented Interface

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots() # Explicit creation
# means are directly creating and managing the objects used to make the plot, rather than relying on Matplotlib to create them automatically behind the scenes.
ax.plot([1, 2, 3], [1, 4, 9]) #this are ploting points
ax.set_title("OO Interface") #set the title in OO interface
ax.set_xlabel("x") #set x label
ax.set_ylabel("y") #set y lable
ax.grid(True) #grid on
plt.show() #display plot
````

# Markers In PLot

You can use the keyword argument marker to emphasize each point with a specified marker:



````python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10]) #we only used y-axis point
#its pylpot that auot create or manage x-axis
#default its start from 0 (on x-axis if not provided)
# it will work likt this example: np.array([0,1,2,3])

plt.plot(ypoints, marker = 'o') #marker can be used to show the marker at edges
plt.show()
````

You can use `ms` the keyword argument markersize or the shorter version, `ms` to set the size of the markers:

````python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20) #if no. of ms increase marker size increase too
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'g', mfc = 'r') #mec:- Marker Edge Color, mfc:- Marker Face Color
# we can wite it like this too:
# mec = '#4CAF50', mfc = '#4CAF50'
# mec = 'hotpink', mfc = 'hotpink' - Or any of the 140 supported color names.
# color name :- https://www.w3schools.com/colors/colors_names.asp
plt.show()
````

# Line Width & Multiple Lines In PyPlot

You can use the keyword argument linewidth or the shorter *`lw`* to change the width of the line.

The value is a floating number, in points: *`linewidth = '20.5'`*

You can plot as many lines as you like by simply adding more *`plt.plot()`* functions:

````python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1,lw='19.2',c='g')
plt.plot(y2)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2) #it will conside x1(x-axis), y1(y-axis), x2(x-axis), y2(y-axis) means two axis total
plt.show()
````

# Matplotlib Labels and Title

With Pyplot, you can use the*` xlabel()`* and *`ylabel()`* functions to set a label for the x- and y-axis.

With Pyplot, you can use the *`title()`* function to set a title for the plot.



````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

You can use the `fontdict` parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels.

You can use the loc parameter in `title()` to position the title.


*   Legal values are: 'left', 'right', and 'center'.
*   Default value is 'center'.



````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20} #passing font family, color, and size (can pass as you want)
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left') #use loc to set location of lable. default center
plt.xlabel("Average Pulse", fontdict = font2) # use fontdict to apply that font style
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
````

# Matplotlib Subplot

With the `subplot()` function you can draw multiple plots in one figure:


- The `subplot()` function takes three arguments that describes the layout of the figure.
- The layout is organized in rows and columns, which are represented by the first and second argument.
- The third argument represents the index of the current plot.


```
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
```
```
plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
```
1. You can draw as many plots you like on one figure, just descibe the number of `rows`, `columns`, and the `index` of the plot.
2. You can add a title to each plot with the `title()` function.
3. You can add a title to the entire figure with the `suptitle()` function.

````python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) # 2-rows, 1-col, 1st plot
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2) #2-rows, 1-col, 2st plot
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP") #adding super title to subplots
plt.show()
````

# ColorMap in Scatter

The **Matplotlib** module has a number of available `colormaps`.

A **colormap** is like a list of colors, where each color has a value that ranges from `0 to 100`.

the `colormap` is called '`viridis`' and as you can see it ranges from 0, which is a purple color, up to 100, which is a yellow color. In below example.

- You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
- In addition you have to create an array with values (from 0 to 100), one value for each point in the scatter plo
- You can include the colormap in the drawing by including the `plt.colorbar()` statement
- find more color map in [Available ColorMaps
](https://www.w3schools.com/python/matplotlib_scatter.asp)

**Example**:

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis') #seting c means color range and its mapping with cmap to viridis

plt.colorbar() #this will show the all the color in range means color bar

plt.show()
````

***Size*** of scatter and its ***opacity***(alpha)

By passing array of sizes and use alpha

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
````

# Matplotlib Histograms

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

The `hist()` function will read the array and produce a histogram

In **Matplotlib**, we use the `hist()` function to create histograms.

The `hist()` function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

````python
import matplotlib.pyplot as plt

# Sample data
ages = [22, 25, 27, 30, 30, 32, 34, 35, 35, 36, 38, 40, 41, 42, 44, 45, 46, 48, 50]

# Create histogram
plt.hist(ages, bins=6, color='skyblue', edgecolor='black')


# Add titles and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
````

**Detailed Info On Histogram**

In Python, `hist()` is a method from **Matplotlib** (`matplotlib.pyplot.hist`) used to create a **histogram** — a type of plot that shows the distribution of a dataset.
- It divides the data into bins (intervals).
- Then it counts how many values fall into each bin.
- Finally, it plots this information as bars.

`plt.hist(data, bins=10, color='blue', edgecolor='black')`

- `data`: A list, NumPy array, or Pandas Series of numbers.
- `bins`: (Optional) Number of intervals (default is 10). Bins are intervals the data is grouped into.
- `color`: Fill color of the bars.
- `edgecolor`: Color of the edges of the bars.

In Above Example:
the range is from 22 to 50.

So each `bin ≈ (50 - 22) / 6 = 28 / 6 ≈ 4.67`,
So the bins might look like:
```
Bin 1: 22.0 – 26.67
Bin 2: 26.67 – 31.33
Bin 3: 31.33 – 36.0
Bin 4: 36.0 – 40.67
Bin 5: 40.67 – 45.33
Bin 6: 45.33 – 50.0
```

How is `Y-axis` (frequency) calculated?

For each bin, Matplotlib counts how many values from the data fall into that interval.

That count becomes the height (Y-axis value) of the bar.

4. Example of counts:
```
Let’s say:

Bin 1: 22, 25 → 2 (only two value rise)
Bin 2: 27, 30, 30 → 3
Bin 3: 32, 34, 35, 35, 36 → 5
Bin 4: 38, 40 → 2
Bin 5: 41, 42, 44, 45 → 4
Bin 6: 46, 48, 50 → 3
```
So your Y-axis (frequencies) will be:

`[2, 3, 5, 2, 4, 3]`

# Matplotlib Pie Charts

With Pyplot, you can use the pie() function to draw pie charts

````python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
````


```
           90
           |
           |
180   ------------   0   <- start
           |
           |
          270

```
**Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:**

The value divided by the sum of all values: `x/sum(x)`

Add arrays of `labels` to the pie chart with the `labels` parameter.

start angle is at the `x-axis`, but you can change the start angle by specifying a startangle parameter.

The `startangle` parameter is defined with an angle in degrees, default angle is 0.

````python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
````

1. **Explode** :- awedges to stand out

- Maybe you want one of the wedges to stand out? The `explode` parameter allows you to do that.

- The `explode` parameter, if specified, and not None, must be an array with one value for each wedge.

- range 0 to 1 (pass array of explode for each).

2. **Shadow** : Add a **`shadow`** to the pie chart by setting the `shadows` parameter to `True`.

3. **Colors** : You can set the `color` of each wedge with the colors parameter.

4. **Legend** : To add a list of explanation for each wedge, use the `legend()` function.
- To add a **`header`** to the `legend`, add the `title` parameter to the `legend` function.

**Example**:

````python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0]

# for example not necessary to include, just 'y'.
plt.pie(y, labels = mylabels, explode = myexplode, colors = mycolors, shadow = True) #we include all like exlpode, colors, shadow.
plt.legend(title = "Four Fruits:") # use title to name the legend
plt.show()
````
